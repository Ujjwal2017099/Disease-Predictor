```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
def days_to_years(days):
    return days/365.25

def bmi_calc(height,weight):
    return weight/(height/100)**2

data = pd.read_csv('cardio_train.csv')
data["age_years"] = days_to_years(data["age"])
data["bmi"] = bmi_calc(data["height"],data["weight"])
```

```python
new_data = df[["bmi","cholesterol" ,"age_years"]]
new_output = df["cardio"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data, train_output, test_output = train_test_split(new_data, new_output, test_size=0.2, random_state=42)
```

```python
clf = RandomForestClassifier(n_estimators=10000,random_state=42)
clf = clf.fit(train_data, train_output)
```

```python
test_prediction = clf.predict(test_data)
```

```python
def rmse(target,prediction):
  return np.sqrt(np.mean(np.square(target-prediction)))
```

```python
rmse(test_output,test_prediction)
```
